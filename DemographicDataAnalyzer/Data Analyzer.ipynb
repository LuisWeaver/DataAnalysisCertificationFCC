##### Qual é a idade média dos homens?

```python
average_age_men = df[df['sex'] == 'Male']['age'].mean().round(1)

average_age_men
```

##### Porcentagem das pessoas que tem um diploma de Bacharel.

```python
# Quantidade de pessoas com diploma de Bacharel
num_bachelors = len(df[df['education'] == 'Bachelors'])
  # num_bachelors = df[df['education'] == 'Bachelors'].shape[0]

num_total = len(df)
  # num_total = df.shape[0]

percentage_bachelors = round(num_bachelors / num_total * 100, 1)

percentage_bachelors
```

##### Qual a porcentagem de pessoas com educação avançada ('Bachelors', 'Masters', ou 'Doctorade') que ganham mais de 50K?

```python
# Quantidade de pessoas com educação avançada e atrasada
higher_education = df[df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])]

lower_education = df[~df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])]
```

```python
# Porcentagem das pessoas com educação avançada que possui um salário >50K
percentage_higher = len(higher_education[higher_education.salary == '>50K'])
  # percentage_higher = len(higher_education[higher_education['salary'] == '>50K'])

higher_education_rich = round(percentage_higher / len(higher_education) * 100, 1)

higher_education_rich
```

```python
# Porcentagem das pessoas com educação atrasada que possui um salário >50K
percentage_lower = len(lower_education[lower_education.salary == '>50K'])
  # percentage_lower = len(lower_education[lower_education['salary'] == '>50K'])

lower_education_rich = round(percentage_lower / len(lower_education) * 100, 1)

lower_education_rich
```

##### Qual o mínimo de horas que alguém trabalha por semana?

```python
# Menor valor de horas presente na tabela
min_works_hours = df['hours-per-week'].min()

min_works_hours
```

```python
# Quantidade de pessoas que trabalham o mínimo de horas semanais
num_min_workers = df[df['hours-per-week'] == min_works_hours]

# Porcentagem dessas pessoas que ganham mais de 50K
rich_percentage = round(len(num_min_workers[num_min_workers.salary == '>50K']) / len(num_min_workers) * 100, 1)

rich_percentage
```

##### Qual país tem a maior porcentagem de pessoas que ganha mais de 50K?

```python
# Distribuição de pessoas por países
country_count = df['native-country'].value_counts()

# Distribuição de pessoas que ganham mais de 50K por países
country_rich_count = df[df['salary'] == '>50K']['native-country'].value_counts()

# País que possui a maior porcentagem de pessoas que ganham mais de 50K
highest_earning_country = (country_rich_count / country_count * 100).idxmax()

print(highest_earning_country)

# Porcentagem de pessoas que ganham mais de 50K no país com maior porcentagem (Iran)
highest_earning_country_percentage = round((country_rich_count / country_count * 100).max(), 1)

highest_earning_country_percentage
```

##### Identificar a ocupação mais popular entre as pessoas que ganham mais de 50K na Índia.

```python
# Pessoas na Índia que ganham mais de 50K
people_of_india_50k = df[(df['native-country'] == 'India') & (df['salary'] == '>50K')]

# Distribuição das ocupações de pessoas que ganham mais de 50K na Índia
occupation_counts = people_of_india_50k['occupation'].value_counts()

# Ocupação com mais pessoas ganhando mais de 50K
top_IN_occupation = occupation_counts.idxmax()

top_IN_occupation
```
